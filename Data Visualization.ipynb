```python
## Plot the Distrubution

data=df['price']
sns.distplot(data, bins=10, hist=True, kde=True, label = 'price')
```

```python
# Heatmap: Correlation between numerical features
plt.figure(figsize=(10, 6))
numerical_cols = df.select_dtypes(include='number')  # only numeric columns
sns.heatmap(numerical_cols.corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
### This kurtosis value is approximately 1.96.Since 1.96 is less than 3, it indicates that the distribution of the price data is platykurtic.
### The distribution of the price column is relatively flat and has lighter tails compared to a normal distribution. 
```

```python
# 1. Scatter Plot: Area vs Price colored by Furnishing Status

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='area', y='price', hue='furnishingstatus', palette='viridis')
plt.title('Area vs Price colored by Furnishing Status')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.legend(title='Furnishing Status')
plt.tight_layout()
plt.show()
```

```python
# 2. Box Plot: Price distribution by Furnishing Status
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='furnishingstatus', y='price', palette='Set2')
plt.title('Price Distribution by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# 3. Bar Plot: Average Price by Number of Bedrooms
avg_price_bedrooms = df.groupby('bedrooms')['price'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_price_bedrooms, x='bedrooms', y='price', palette='coolwarm')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

```python
# 4. Pairplot: Selected numerical features
selected_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
sns.pairplot(df[selected_cols], diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot of Selected Numerical Features', y=1.02)
plt.tight_layout()
plt.show()
```
