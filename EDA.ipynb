```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(bins=20, figsize=(16, 12), color='lightblue', edgecolor='black')
plt.suptitle("Distribution of Features", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
plt.figure(figsize=(16, 12))
for i, col in enumerate(df.columns):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y=df[col], color='salmon')
    plt.title(col)
plt.tight_layout()
plt.suptitle("Box Plots for Outlier Detection", fontsize=18, y=1.02)
plt.show()
```

```python
plt.figure(figsize=(14, 10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
if 'target' in df.columns:
    plt.figure(figsize=(16, 10))
    selected_features = df.columns[df.columns != 'target']
    for i, col in enumerate(selected_features[:8]):
        plt.subplot(2, 4, i+1)
        sns.boxplot(x='target', y=col, data=df, palette='Set2')
        plt.title(f"{col} by Target")
    plt.tight_layout()
    plt.suptitle("Feature Distributions by Heart Disease Presence", fontsize=18, y=1.02)
    plt.show()
```
